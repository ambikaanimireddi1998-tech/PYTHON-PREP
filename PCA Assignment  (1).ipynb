### Task 2: Dimensionality Reduction with PCA

```python
# 1. Standardize the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
# 2. Apply PCA
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Explained variance to determine number of components
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

plt.figure(figsize=(8,5))
plt.plot(range(1,len(explained_variance)+1), cumulative_variance, marker='o', linestyle='--')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot")
plt.grid()
plt.show()
```

```python
# Choosing number of components (e.g., 95% variance)
n_components = np.argmax(cumulative_variance >= 0.95) + 1
print("Number of components to explain 95% variance:", n_components)

pca = PCA(n_components=n_components)
X_pca_final = pca.fit_transform(X_scaled)
```

### Task 3: Clustering with Original Data

```python
# 1. Apply K-means clustering
from sklearn.cluster import KMeans
kmeans_orig = KMeans(n_clusters=3, random_state=42)
clusters_orig = kmeans_orig.fit_predict(X_scaled)
```

```python
# 2. Visualize clustering (using first two features)
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=clusters_orig, palette='Set1')
plt.title("K-means Clustering on Original Data")
plt.show()
```

```python
# 3. Evaluate clustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
sil_score_orig = silhouette_score(X_scaled, clusters_orig)
db_score_orig = davies_bouldin_score(X_scaled, clusters_orig)
print("Original Data - Silhouette Score:", sil_score_orig)
print("Original Data - Davies-Bouldin Index:", db_score_orig)
```

### Task 4: Clustering with PCA Data

```python
# 1. Apply K-means on PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca_final)
```

```python
# 2. Visualize clustering (first two principal components)
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca_final[:,0], y=X_pca_final[:,1], hue=clusters_pca, palette='Set1')
plt.title("K-means Clustering on PCA Data")
plt.show()
```

```python
# 3. Evaluate clustering
sil_score_pca = silhouette_score(X_pca_final, clusters_pca)
db_score_pca = davies_bouldin_score(X_pca_final, clusters_pca)
print("PCA Data - Silhouette Score:", sil_score_pca)
print("PCA Data - Davies-Bouldin Index:", db_score_pca)
```

### Task 5: Comparison and Analysis

```python
print("\nComparison of Clustering:")
print(f"Silhouette Score - Original: {sil_score_orig:.3f}, PCA: {sil_score_pca:.3f}")
print(f"Davies-Bouldin Index - Original: {db_score_orig:.3f}, PCA: {db_score_pca:.3f}")
```

### Task 6: Conclusion and Insights

```python
print("\nConclusion & Insights:")
print("- PCA reduced the dimensionality while retaining ~95% variance.")
print("- Clustering on PCA data is faster and often similar to original data clustering.")
print("- If the original data has many features, PCA helps to simplify visualization and computation.")
print("- Trade-off: Some minor information may be lost in PCA, which might slightly affect clustering quality.")
```

```python
# 1. Key Findings and Insights

#The original dataset had multiple features, some of which were correlated.

#PCA reduced the dataset to fewer components (while retaining ~95% of variance), simplifying the structure.

#Clustering on both the original dataset and PCA-transformed data produced similar groupings, indicating that PCA preserved the main patterns in the data.

#Visualizations showed that PCA made it easier to separate clusters clearly in 2D or 3D space.

#PCA also improved computational efficiency for clustering, especially with high-dimensional datasets.
```

```python
# 2. Practical Implications of Using PCA and Clustering

#PCA is valuable when dealing with high-dimensional datasets, as it reduces complexity while retaining most of the data’s information.

#It helps remove noise and correlations among features, making clustering and other analyses more robust.

#Clustering can be applied either to the original data or PCA-transformed data. Using PCA first allows faster computation and better visualization without significantly affecting clustering quality.

#Combining PCA with clustering is particularly useful in exploratory data analysis, as it provides insights into the natural grouping of data in fewer dimensions.
```

```python
# 3. Recommendations

#Use PCA when: high-dimensional data, correlated features, or visualization in 2D/3D is needed.

#Use clustering on original data when: features are few and all information is important.

#Combine PCA and clustering for large, complex datasets to balance accuracy, interpretability, and computational efficiency.
```
