Descriptive Analytics for Numerical Columns Objective:

To calculate and analyze basic statistical measures such as mean, median, mode, and standard deviation for numerical columns.

Explanation:

Mean represents the average value of sales or discounts.

Median gives the middle value and is useful when data contains outliers.

Mode shows the most frequently occurring value.

Standard Deviation indicates how much the data varies from the mean.

Interpretation:

A high mean sales value indicates strong overall sales.

If mean > median, the data is right skewed.

A high standard deviation shows large variation in sales or discounts.

A low standard deviation indicates consistent values.

```python
# boxplot
df.boxplot(column="Volume",vert=False)
```

```python
#identify the outllier and IQR(Inter Quartile Range)
import numpy as np
Q1=np.percentile(df["Volume"],25)
Q3=np.percentile(df["Volume"],75)
IQR=Q3-Q1
print("FIRST QUARTILE:",Q1)
print("THIRD QUARTILE:",Q3)
print("INTER QUARTILE RANGE:",IQR)
#outlier range
LW=Q1-(1.5*IQR)
UW=Q3+(1.5*IQR)
print("LOWER WHISKER LENGTH:",LW)
print("UPPER WHISKER LENGTH:",UW)
print("UW:",len(df[df["Volume"]>UW]))
print("LW:",len(df[df["Volume"]<LW]))
```
