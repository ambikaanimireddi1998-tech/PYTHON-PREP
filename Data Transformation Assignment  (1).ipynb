```python
# Numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

# Impute numerical features
for col in num_cols:
    df[col].fillna(df[col].median())

# Impute categorical features
for col in cat_cols:
    df[col].fillna(df[col].mode()[0])
```

```python
print("Skewness before:", df['capital_gain'].skew())

# Log transformation
df['capital_gain_log'] = np.log1p(df['capital_gain'])

print("Skewness after:", df['capital_gain_log'].skew())
```

```python
from sklearn.preprocessing import StandardScaler
scaler_std = StandardScaler()
df_standard = df.copy()

df_standard[num_cols] = scaler_std.fit_transform(df_standard[num_cols])
```

```python
from sklearn.ensemble import IsolationForest
iso = IsolationForest(contamination=0.05, random_state=42)

outliers = iso.fit_predict(df[num_cols])

# Keep only normal points
df = df[outliers == 1]

print("Shape after removing outliers:", df.shape)
```
